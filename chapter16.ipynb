16-1 [ ] 이용하기

```python
#조건을 충족하는 행 추출하기
import pandas as pd
df_raw = pd.read_csv('exam.csv')
df = df_raw.head(10)
df
```

```python
#nclass 추출
df['nclass']
```

```python
#nclass가 1인지 확인
df['nclass'] == 1
```

```python
#nclass가 1이면서 수학 점수가 50점 이상
df[(df['nclass'] == 1) & (df['math'] >= 50)]
```

```python
#열 추출
df['id']
```

```python
#여러 변수
df[['id', 'nclass']]
```

```python
#열이 1개일 때 데이터 프레임 자료 구조 유지하기
df[['id']]
```

```python
#조건을 충족하는 행에서 열 추출하기
#nclass가 1인 행의 math열
df[df['nclass'] == 1]['math']
```

[ ]를 pandas 함수와 함께 사용하기

```python
#pandas 함수만 사용
df.groupby('nclass') \
  .agg(math = ('math', 'mean'),
       english = ('english', 'mean'))
```

```python
#pandas 함수와 [ ] 함께 사용
df.groupby('nclass')[['math', 'english']].mean()
```

**16-2 df.loc[ ] 이용하기**

```python
#인덱스 활용하기
df = pd.DataFrame({'var1' : [1, 2, 3],
                   'var2' : [4, 5, 6]})
df
```

```python
#인덱스 문자열 지정하기
df = pd.DataFrame({'var1' : [1, 2, 3],
                   'var2' : [4, 5, 6]},
                   index = ['kim', 'lee', 'park'])
df
```

```python
#인덱스 문자열을 지정해 행 추출하기
df.loc['kim']
```

```python
#여러 행 추출하기
df.loc[['kim', 'park']]
```

```python
#인덱스 문자열을 지정해 열 추출하기
#lee행의 var1열 추출
df.loc['lee', 'var1']
```

```python
#모든 행에 대한 열 추출
df.loc[:, 'var1']
```

```python
#인덱스 번호를 지정해 행 추출하기
df = df_raw.copy()
df
```

```python
#0행 추출
df.loc[0]
```

```python
#연속된 행 추출하기
df.loc[7:9]
```

```python
#인덱스 번호로 행은 추출할 수 있지만, 열은 추출할 수 없다
df.loc[:, 2]
```

```python
#nclass가 1이면 추출
df.loc[df['nclass'] == 1]
```

```python
#nclass가 1인 행의 math, english열 추출
df.loc[df['nclass'] == 1,['math', 'english']]
```

```python
#위와 같은 방법
df[df['nclass'] == 1][['math', 'english']]
```

**16-3 df.iloc[ ] 이용하기**

```python
#0행 추출
df.iloc[0]
```

```python
#연속된 행 추출
#1~4행 출력
df.iloc[1:5]
```

```python
#인덱스 문자열만 있어도 인덱스 번호로 행 추출 가능
df = pd.DataFrame({'var1' : [1, 2, 3],
                   'var2' : [4, 5, 6]},
                   index = ['kim', 'lee', 'park'])
df.iloc[0]
```

```python
#인덱스 번호를 지정해 열 추출하기
#모든 행의 1, 3열 추출
df.iloc[:, [1, 3]]
```

```python
#0, 1행의 2, 3열 추출
df.iloc[[0, 1], [2, 3]]
```
