**04-2 데이터 프레임 만들기**

데이터를 가공할 때 사용하는 pandas 패키지


```python
import pandas as pd
```

pandas 패키지의 DataFrame() 함수 사용하기

```python
df = pd.DataFrame({'name' : ['김지훈', '이유진', '박동현', '김민지'],
                   'english' : [90, 80, 60, 70],
                   'math' : [50, 60, 100, 20]})
df
```

특정 변수의 값 출력

```python
df['english']
```

```python
sum(df['english'])
```

변수의 값으로 평균 구하기

```python
sum(df['math']) / 4
```

연습문제 : 데이터 프레임 만들어 출력하기

```python
df = pd.DataFrame({'제품' : ['사과', '딸기', '수박'],
                  '가격' : [1800, 1500, 3000],
                  '판매량' : [24, 38, 13]})
df
```

```python
sum(df['가격']) / 3
```

```python
sum(df['판매량']) / 3
```

**04-3 외부 데이터 이용하기**

pandas의 read_excel()을 이용해 엑셀 파일 불러오기 (경로 확인)

```python
df_exam = pd.read_excel('excel_exam.xlsx')
df_exam
```

```python
sum(df_exam['english']) / 20
```

len()으로 값의 개수(행의 개수) 구하기

```python
sum(df_exam['science']) / len(df_exam)
```

sheet_name 이용해서 시트의 데이터 불러오기

```python
df_exam = pd.read_excel('excel_exam.xlsx', sheet_name = 'Sheet1')
df_exam
```

```python
df_exam = pd.read_excel('excel_exam.xlsx', sheet_name = 0)
df_exam
```

read_csv() 이용해서 CSV 파일 불러오기

```python
df_csv_exam = pd.read_csv('exam.csv')
df_csv_exam
```

데이터 프레임을 CSV 파일로 저장하기

```python
df_midterm = pd.DataFrame({'english' : [90, 80, 60, 70],
                           'math' : [50, 60, 100, 20],
                           'nclass' : [1, 1, 2, 2]})
df_midterm
```

to_csv() 이용해서 CSV 파일로 저장하기

```python
df_midterm.to_csv("output_newdata.csv")
```
