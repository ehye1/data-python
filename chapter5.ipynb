조건문을 활용해 파생변수 만들기

> 연비가 기준값을 넘으면 합격, 넘지 못하면 불합격



```python
mpg['total'] = (mpg['cty'] + mpg['hwy']) / 2
mpg['total'].describe()
```

```python
mpg['total'].plot.hist()
```

-> 요약 통계량과 히스토그램을 본 결과 total 연비의 평균과 중간값이 약 20이다

=> 기준값을 20으로 정한다

NumPy 패키지의 where()을 이용하여 합격 판정 변수 만들기

```python
import numpy as np
```

```python
mpg['test'] = np.where(mpg['total'] >= 20, 'pass', 'fail')
mpg.head()
```

```python
mpg['test'].value_counts()
```

```python
count_test = mpg['test'].value_counts()
count_test.plot.bar()
```

축 이름 회전

```python
count_test.plot.bar(rot = 0)
```

중첩 조건문 만들기

```python
mpg['grade'] = np.where(mpg['total'] >= 30, 'A',
                        np.where(mpg['total'] >= 20, 'B', 'C'))
mpg.head()
```

```python
count_grade = mpg['grade'].value_counts()
count_grade
```

```python
count_grade.plot.bar(rot = 0)
```

```python
count_grade = count_grade.sort_index()
count_grade
```

```python
count_grade.plot.bar(rot = 0)
```

버티컬 바를 사용하여 목록에 해당하는 행으로 변수 만들기

```python
mpg['size'] = np.where((mpg['category'] == 'compact') |
                       (mpg['category'] == 'subcompact') |
                       (mpg['category'] == '2seater'),
                       'small', 'large')
mpg['size'].value_counts()
```

isin()을 사용하여 코드 간편하게 작성하기

```python
mpg['size'] = np.where(mpg['category'].isin(['compact','subcompact','2seater']),
                       'small', 'large')
mpg['size'].value_counts()
```
