**06-5 파생변수 추가하기**

```python
#assign()을 이용하여 데이터에 파생변수 만들기
exam.assign(total = exam['math'] + exam['english'] + exam['science'])
```

```python
#df.assign()에 np.where() 적용하기
import numpy as np
exam.assign(test = np.where(exam['science'] >= 60, 'pass', 'fail'))
```

```python
#lambda 이용하여 데이터 프레임명 줄여 쓰기
exam.assign(new = lambda x: x['math'] + x['science'])
```

**06-6 집단별로 요약하기**

```python
#agg()로 요약통계량 구하기
#요약값을 할당할 변수명 = 요약하는데 사용할 변수,함수
exam.agg(mean_math = ('math', 'mean'))
```

```python
#groupby()로 집단별 요약 통계량 구하기
exam.groupby('nclass') \
    .agg(mean_math = ('math', 'mean'))
```

**06-7 데이터 합치기**

```python
#가로로 합치기
test1 = pd.DataFrame({'id' : [1, 2, 3, 4, 5],
                      'midtern' : [60, 80, 70, 90, 85]})
test1
```

```python
test2 = pd.DataFrame({'id' : [1, 2, 3, 4, 5],
                      'midtern' : [70, 83, 65, 95, 80]})
test2
```

```python
total = pd.merge(test1, test2, how = 'left', on = 'id')
total
```

```python
#세로로 합치기
#두 데이터의 변수명이 같아야 한다
group_a = pd.DataFrame({'id' : [1, 2, 3, 4, 5],
                        'test' : [60, 80, 70, 90, 85]})
group_b = pd.DataFrame({'id' : [6, 7, 8, 9, 10],
                        'test' : [70, 83, 65, 95, 80]})
group_all = pd.concat([group_a, group_b])
group_all
```
